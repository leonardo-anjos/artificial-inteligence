# Análise de Sentimento e Emoção

Análise de Sentimentos, também conhecida como mineração de opinião ou Emotion A.I. é o processamento da linguagem natural e análise de texto para extrair o conteúdo emocional e as opiniões expressas em um texto. 

Pode ser usado para entender atitudes, ideias ou pensamentos expressos no texto. Muito utilizado para determinar se a atitude do autor do texto em relação a um tópico ou produto é positiva, negativa ou neutra.

## Sentimento

Está relacionada aos aspectos abaixo:

- **Positivo**. Valor baixo indica que é improvável que seja percebido como positivo. Alto valor indica muito provável que seja percebido como positivo.


- **Negativo**. Valor baixo indica que é improvável que seja percebido como negativo. Alto valor indica muito provável que seja percebido como negativo.


## Emoção

Está relacionada às emoções do tipo:

- **Raiva** *(Anger)*

  Probabilidade de escritor ser percebido com raiva. Valor baixo indica que é improvável que seja percebido como irritado. Alto valor indica muito provável que seja percebido como irritado


- **Medo** *(Fear)*

  Probabilidade de escritor ser percebido como assustado. Valor baixo indica improvável que seja percebido como medroso. Alto valor, muito provavelmente percebido como assustado.


- **Alegria** *(Joy)*

  Alegria ou felicidade tem nuances de prazer e satisfação. Há uma sensação de bem-estar, paz interior, amor, segurança e contentamento.
  

- **Tristeza** *(Sadness)*

  Probabilidade de escritor ser percebido como triste. Valor baixo, dificilmente percebido como triste. Alto valor muito provável de ser percebido como triste.


- **Analítico** *(Analytical)*

  Raciocínio e atitude analítica de um escritor sobre as coisas. Maior valor, mais provável de ser percebido como intelectual, racional, sistemático, sem emoção ou impessoal.


- **Confiante** *(Confident)*

  Grau de certeza de um escritor. Maior valor, mais provável de ser percebido como assegurado, coletado, esperançoso ou egoísta.


- **Tentativa** *(Tentative)*

  Grau de inibição de um escritor. Maior valor, mais provável de ser percebido como questionável, duvidoso, limitado ou discutível.

# Preparação para uso dos Notebooks

Utilizamos o Anaconda para gerenciar os ambientes Python.

Uma vez instalado o Anaconda 3.5+ execute o comando abaixo para criar o ambiente:

```
conda create -n pln_ase python==3.6
```

Uma vez criado o ambiente, ative-o com o comando abaixo:

```
source activate pln_ase
```
 
Instale o jupyter com o comando abaixo:
