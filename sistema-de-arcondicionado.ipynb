# inteligência artificial

Sistema de controle de temperatura de ar-condicionado com 5 níveis de fuzzy

Este sistema faz o ajuste automático do ar-condicionado comparado com a temperatura do ambiente




**Variáveis linguísticas e termos:**

Entradas e saídas das variáveis

Temperatura(t) = {muito frio, frio, normal, quente, muito quente}

**Funções de pertinencia:**

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

temp = np.arange(0, 40, 1)

temp_mf = fuzz.trimf(temp, [0, 0, 5])
temp_f = fuzz.trimf(temp, [0, 5, 10])
temp_n = fuzz.trimf(temp, [5, 10, 10])
temp_q = fuzz.trimf(temp, [0, 0, 5])
temp_mq = fuzz.trimf(temp, [0, 5, 10])
```

**Visualização das funções de membro:**

```python
fig, (ax0) = plt.subplots(nrows=1)
ax0.plot(temp, temp_mf, 'b', linewidth=1.5, label='Bad')
ax0.plot(temp, temp_f, 'g', linewidth=1.5, label='Decent')
ax0.plot(temp, temp_n, 'r', linewidth=1.5, label='Great')
ax0.plot(temp, temp_q, 'c', linewidth=1.5, label='Decent')
ax0.plot(temp, temp_mq, 'm', linewidth=1.5, label='Great')
ax0.set_title('Temperatura')
ax0.legend()
```
